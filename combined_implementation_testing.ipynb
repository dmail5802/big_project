## Example Raspberry Pi workflow

This example shows how the finished notebook would read the Pi sensors and have the Pi speak a message if a Raspberry Pi were available.

### establish connection 

```python
# Connect to the Pi
pi = RaspberryPiSensor(
    host=PI_HOST,
    username=PI_USERNAME,
    password=PI_PASSWORD,
    prompt_for_password=False,
)
```

### read the sensors

```python
# Read the sensors
{'light':pi.get_light_reading(light_sensor_port=0).get('value'),
 'temperature': pi.get_temperature_reading(dht_sensor_port=5, dht_sensor_type=0).get('temperature_c')}
```

### create raspi message

```python
# Build a message for the Pi to speak
(message := (
    f"The light level is {num2words(pi.get_light_reading(light_sensor_port=0).get('value'))} and "
    f"the temperature is {num2words(pi.get_temperature_reading(dht_sensor_port=5, dht_sensor_type=0).get('temperature_c'))} degrees Celsius."
))
```

### create a voice file from the message saved in voice modelartifacts-> synthesized_audio

```python
# Create a voice file locally, then copy or place it on the Pi if needed
voice_service = VoiceFileService(project_root=directory)
voice_file_path = voice_service.create_voice_file(
    text=message,
    output_filename="pi_status_message.wav",
)
```
